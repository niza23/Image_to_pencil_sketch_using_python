```python
#Importing required libraries
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/drive/MyDrive/pic.jpg")
cv2_imshow(image)
```

```python
#Converting a BGR image into RGB image
rgbimg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(rgbimg)
```

```python
#Converting a RGB image into gray_scale image
grayimg = cv2.cvtColor(rgbimg, cv2.COLOR_RGB2GRAY)
cv2_imshow(grayimg)
```

```python
#Inverting the image
invertimg = cv2.bitwise_not(grayimg)
cv2_imshow(invertimg)
```

```python
# Blurring the image
blurimg= cv2.GaussianBlur(invertimg, (15, 15), sigmaX=0, sigmaY=0)
cv2_imshow(blurimg)
```

```python
#Pencil Sketch
pencil_sketch = cv2.divide(grayimg, 255 - blurimg, scale=256)
```

```python

cv2_imshow(pencil_sketch)
```
